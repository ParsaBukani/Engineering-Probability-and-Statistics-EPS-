## Section 1:

```r
n <- 10^6
uni_samples <- runif(n, min = 0, max = 1)

par(bg = "white")
hist(uni_samples, breaks = 100, probability = TRUE, col = 'lightgreen',
     main = 'Uniform Distribution',
     xlab = 'Value', ylab = 'Density')
```

## Section 2:

```r
exponential_samples <- -2 * log(uni_samples)
```

## Section 3:

```r
exponential_density <- function(x, lambda = 0.5) {
  return (dexp(rate = lambda, x = x))
}

par(bg = "white")
hist(exponential_samples, breaks = 100, probability = TRUE, main = "Empirical vs Theoretical Exponential Distribution",
     xlab = "Values", col = "lightblue", border = "black")

x_values <- seq(min(exponential_samples), max(exponential_samples), length.out=100)
lines(x_values, exponential_density(x_values), col = 'red', lwd = 2)

legend("topright", legend = c("Empirical Distribution", "Theoretical Distribution"),
       col = c("lightblue", "red"), lwd = 2)
```

**Report**: By comparing the histogram with the theoretical density curve, we can observe that the empirical distribution closely follows the theoretical exponential distribution, confirming the correctness of the transformation.

## Section 4:

```r
U1 <- runif(n, min = 0, max = 1)
U2 <- runif(n, min = 0, max = 1)
```

## Section 5:

```r
R <- sqrt(-2 * log(U1))
theta <- 2 * pi * U2
Z1 <- R * cos(theta)
Z2 <- R * sin(theta)
```

## Section 6:

```r
std_normal_density <- function(x) {
  return(dnorm(x))
}

z1_values <- seq(min(Z1), max(Z1), length.out = 1000)
z2_values <- seq(min(Z2), max(Z2), length.out = 1000)

par(mfrow = c(1, 2), bg = "white")

hist(Z1, breaks = 100, probability = TRUE, main = "Standard Normal Distribution Z1",
     xlab = "Values", col = "lightblue", border = "black")
lines(z1_values, std_normal_density(z1_values), col = 'red', lwd = 2)
legend("topleft", legend = c("Empirical Distribution", "Theoretical Density"), col = c("lightblue", "red"), lwd = 2, cex = 0.7, bty = "n")

hist(Z2, breaks = 100, probability = TRUE, main = "Standard Normal Distribution Z2",
     xlab = "Values", col = "lightblue", border = "black")
lines(z2_values, std_normal_density(z2_values), col = 'red', lwd = 2)
legend("topleft", legend = c("Empirical Distribution", "Theoretical Density"), col = c("lightblue", "red"), lwd = 2, cex = 0.7, bty = "n")

```

**Report:**
By overlaying the theoretical density curves on the histograms, we can visually confirm that the empirical distributions of Z1 and Z2 closely match the standard normal distribution.
This practical example demonstrates how the Box-Muller transform successfully converts uniform random variables into standard normal variables.
