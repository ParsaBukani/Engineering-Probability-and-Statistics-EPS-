# Part 1:

> ### **Poisson distribution:**
*  #### **Mean:**
$$ \mu = \lambda = 10 $$

* #### **Variance:**
$$ \sigma^2 = \lambda = 10 $$

> ### **Exponential Distribution:**
* #### **Mean:**
$$ \mu = \frac{1}{\lambda} = \frac{1}{(1/2)} = 2 $$
* #### **Variance:**
$$ \sigma^2 = \frac{1}{\lambda^2} = \frac{1}{(1/2)^2} = 4 $$

> ### **Geometric Distribution:**
* #### **Mean:**
$$ \mu = \frac{1}{p} = \frac{1}{(1/5)} = 5 $$
* #### **Variance:**
$$ \sigma^2 = \frac{1 - p}{p^2} = \frac{1 - (1/5)}{(1/5)^2} = 20 $$

# Part 2:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import random
np.random.seed(501)

sample_sizes = [30, 300, 3000]
repetitions = 1000
```

> ## Poisson distribution:

```python
plt.figure(figsize=(12, 4))
parameter = 10
means_po = []

for i, size in enumerate(sample_sizes):
    means = []
    for _ in range(repetitions):
        samples = np.random.poisson(lam=parameter, size=size)
        means.append(np.mean(samples))
    
    plt.subplot(1, 3, i + 1)
    plt.hist(means, bins=50)
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('Frequency')
    means_po.append(means)

plt.suptitle(f'Sample Means of Poisson Distribution')
plt.show()
```

> ## Exponential Distribution:

```python
plt.figure(figsize=(12, 4))
parameter = 1/2

means_ex = []
for i, size in enumerate(sample_sizes):
    means = []
    for _ in range(repetitions):
        samples = np.random.exponential(scale= 1/parameter, size=size)
        means.append(np.mean(samples))
    
    plt.subplot(1, 3, i + 1)
    plt.hist(means, bins=50)
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('Frequency')
    means_ex.append(means)

plt.suptitle(f'Sample Means of Exponential Distribution')
plt.show()
```

> ## Geometric Distribution:

```python
plt.figure(figsize=(12, 4))
parameter = 1/5

means_go = []
for i, size in enumerate(sample_sizes):
    means = []
    for _ in range(repetitions):
        samples = np.random.geometric(p=parameter, size=size)
        means.append(np.mean(samples))
    
    plt.subplot(1, 3, i + 1)
    plt.hist(means, bins=50)
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('Frequency')
    means_go.append(means)

plt.suptitle(f'Sample Means of Geometric Distribution')
plt.show()
```

# Part 3:

> ## Poisson distribution:

```python
mean_of_means_po = []
std_of_means_po = []

for i, size in enumerate(sample_sizes):        
        mean = np.mean(means_po[i])
        std = np.std(means_po[i], ddof=1)
        
        print(f'Sample Size: {size}')
        print(f'Mean of Means: {mean}')
        print(f'Standard Deviation of Means: {std}\n')

        mean_of_means_po.append(mean)
        std_of_means_po.append(std)
```

> ## Exponential Distribution:

```python
mean_of_means_ex = []
std_of_means_ex = []

for i, size in enumerate(sample_sizes):        
        mean = np.mean(means_ex[i])
        std = np.std(means_ex[i], ddof=1)
        
        print(f'Sample Size: {size}')
        print(f'Mean of Means: {mean}')
        print(f'Standard Deviation of Means: {std}\n')

        mean_of_means_ex.append(mean)
        std_of_means_ex.append(std)
```

> ## Geometric Distribution:

```python
mean_of_means_go = []
std_of_means_go = []

for i, size in enumerate(sample_sizes):        
        mean = np.mean(means_go[i])
        std = np.std(means_go[i], ddof=1)
        
        print(f'Sample Size: {size}')
        print(f'Mean of Means: {mean}')
        print(f'Standard Deviation of Means: {std}\n')

        mean_of_means_go.append(mean)
        std_of_means_go.append(std)
```

As the sample size increases, the following changes are observed:

1. **Mean of Means:** The mean of the sample means tends to get closer to the true population mean. This is due to the Law of Large Numbers, which states that as the sample size increases, the sample mean converges to the population mean.

2. **Standard Deviation of Means:** The standard deviation of the sample means (also known as the standard error) decreases as the sample size increases. This is because larger samples tend to provide more accurate estimates of the population mean, reducing the variability of the sample means.

# Part 4:

> ## Poisson distribution:

```python
plt.figure(figsize=(12, 3))

for i, size in enumerate(sample_sizes):        
    plt.subplot(1, 3, i + 1)
    plt.hist(means_po[i], bins=50, density=True)
    
    x = np.linspace(8, 12, 100)
    p = stats.norm.pdf(x, mean_of_means_po[i], std_of_means_po[i])
    plt.plot(x, p, 'r', linewidth=2)
    
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('density')
```

> ## Exponential Distribution:

```python
plt.figure(figsize=(12, 3))

for i, size in enumerate(sample_sizes):        
    plt.subplot(1, 3, i + 1)
    plt.hist(means_ex[i], bins=50, density=True)
    
    x = np.linspace(0, 4, 100)
    p = stats.norm.pdf(x, mean_of_means_ex[i], std_of_means_ex[i])
    plt.plot(x, p, 'r', linewidth=2)
    
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('density')
```

> ## Geometric Distribution:

```python
plt.figure(figsize=(12, 3))

for i, size in enumerate(sample_sizes):        
    plt.subplot(1, 3, i + 1)
    plt.hist(means_go[i], bins=50, density=True)
    
    x = np.linspace(2, 9, 100)
    p = stats.norm.pdf(x, mean_of_means_go[i], std_of_means_go[i])
    plt.plot(x, p, 'r', linewidth=2)
    
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('density')
```

**Observations:**

* Overlaying Normal Distribution: As the sample size increases, the histograms of the sample means become more closely aligned with the normal distribution. This is evident from the overlay of the normal distribution curve (red line) on the histograms, which fits better as the sample size grows.

* Convergence to Normal Distribution: With increasing sample size, the distribution of the sample means will tend to a normal distribution. This is due to the Central Limit Theorem.

* Parameters of the Normal Distribution: The parameters of the normal distribution (mean and standard deviation) are dependent on the original distribution's parameters:

    * Mean: The mean of the normal distribution is the same as the mean of the original distribution.
    * Standard Deviation: The standard deviation of the normal distribution (standard error) is the standard deviation of the original distribution divided by the square root of the sample size: $$\text{Standard Error} = \frac{\sigma}{\sqrt{n}}$$

# Part 5:

```python
plt.figure(figsize=(14, 4))

def distributions(distribution, size):
    if distribution == 'Poisson':
        return np.random.poisson(lam=10, size=size),
    elif distribution == 'Exponential':
        return np.random.exponential(scale=2, size=size)
    elif distribution == 'Geometric':
        return np.random.geometric(p=0.2, size=size)


mean_of_means_xx = []
for i, size in enumerate(sample_sizes):
    means = []
    for _ in range(repetitions):
        distribution = random.choice(['Poisson', 'Exponential', 'Geometric'])
        samples = distributions(distribution, size)
        means.append(np.mean(samples))
    
    mean_of_means = np.mean(means)
    std_of_means = np.std(means, ddof=1)

    
    plt.subplot(1, 3, i + 1)
    plt.hist(means, bins=50, density=True)
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('density')

    print(f'Sample Size: {size}') 
    print(f'Mean of Sample Means: {mean_of_means}') 
    print(f'Standard Error of Sample Means: {std_of_means}\n')

    mean_of_means_xx.append((means, mean_of_means, std_of_means))
    

plt.suptitle(f'Histogram of the Sample Means of Random Distributions')
plt.show()
        
```

We expect the distribution of sample means to be normal based on the Central Limit Theorem, which states that the means of sufficiently large samples from any distribution will approximate a normal distribution. However, it is not normal in this case because _the samples are not independent and identically distributed (i.i.d.)_. Each sampling randomly selects a different distribution, violating the key condition for the CLT. Thus, the resulting distribution does not converge to normality.

```python
plt.figure(figsize=(12, 3))

for i, size in enumerate(sample_sizes):        
    plt.subplot(1, 3, i + 1)
    plt.hist(mean_of_means_xx[i][0], bins=50, density=True)
    
    x = np.linspace(2, 9, 100)
    p = stats.norm.pdf(x, mean_of_means_xx[i][1], mean_of_means_xx[i][2])
    plt.plot(x, p, 'r', linewidth=2)
    
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean')
    plt.ylabel('density')
```

* The empirical results from Part 5 show that the distributions of sample means are not normal because the samples are not i.i.d (as discussed before). Another obvious reason is that the Standard Errors for different sample sizes are not following a specific pattern and have fluctuations.

Since the samples in Part 5 were not independent, the distributions obtained in Part 4 cannot be considered jointly normal. Joint normality requires both the condition of individual normality and independence, and these conditions were not met in our observations.

> **Lastly:**

The Central Limit Theorem (CLT) does hold for all distributions, given that certain conditions are met. Specifically, the samples must be independent and identically distributed (i.i.d.), and the sample size should be sufficiently large.
However, if these conditions are not met, as observed in Part 5 where the samples were not i.i.d., the distributions of sample means will not be jointly normal.
